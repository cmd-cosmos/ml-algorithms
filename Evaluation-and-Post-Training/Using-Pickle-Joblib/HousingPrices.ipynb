```python
'''
Author: Premanshu Bhagat
Date: 09/07/2025
Predict housing prices given area in square feet.

Exporting model using joblib or pickle
'''

import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Housing.csv')
df.head()
col_remove = df.columns[2:]
newdf = df.drop(columns=col_remove, axis=1)
newdf.head(20)
```

```python
plt.figure()
plt.xlabel('Area (sq.ft.)')
plt.ylabel('Price (USD)')
plt.scatter(newdf.area, newdf.price)
plt.show()
```

```python
reg =  LinearRegression()
reg.fit(newdf[['area']], newdf['price']) # (independent var, target)
```

```python
reg.predict([[7420]])
```

```python
# Pickle module allows for serialization of a python object to a file

import pickle

with open('model_pickle','wb') as f: # we write the file as binary
    pickle.dump(reg,f)               # dump the model reg into file f
```

```python
# now we do not need to retrain the model in order to use it for predictions

with open('model_pickle','rb') as f:
    model = pickle.load(f)           # now pretrained reg model has been dumped into the model obj

# using the loaded model to make predicitons

model.predict([[7420]]) # same output as before.
```

```python
# joblib model dump
# it is more efficient when used for models with large np arrays

import joblib

joblib.dump(reg, 'joblib_model.pkl')
```

```python
joblibModel = joblib.load('joblib_model.pkl')
joblibModel.predict([[7420]]) # Same value predicted
```
