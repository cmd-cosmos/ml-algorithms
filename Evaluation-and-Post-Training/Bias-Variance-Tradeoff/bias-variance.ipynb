# Bias
- difference between the prediction of the values by the Machine Learning model and the correct value.
- essentially a measure of how accurately a model can capture patterns in the training dataset.

# Variance
- The variability of model prediction for a given data point which tells us the spread of our data is called the variance of the model.
- model with high variance has a very complex fit to the training data.
- test error value dependent.
- such models perform very well on training data but have high error rates on test data.
