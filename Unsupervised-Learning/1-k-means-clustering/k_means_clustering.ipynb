# k-means clustering

- unsupervised learning algorithm  ---> target var unknown.
- find patterns in data and make predictions based on that.
- k-means ---> k clusters from the data ---> k centroids for the clusters
          ---> k is passed as a parameter to the algorithm

- Algorithm K-Means Clustering:

1. Initialize centroids
   - Randomly select k data points from the dataset as initial centroids.

2. Repeat until convergence:
   a. Assignment step:
      - For each data point in the dataset:
        i.  Calculate the distance between the data point and each centroid.
        ii. Assign the data point to the nearest centroid.
   
   b. Update step:
      - For each centroid:
        i.  Calculate the new centroid by taking the mean of all data points assigned to it.

3. Convergence criteria:
   - Check if the centroids have stopped moving (i.e., the changes in centroid positions are below a certain threshold).
   - If centroids have converged, terminate the algorithm.
   - If not, repeat steps 2a and 2b.

End Algorithm



# Elbow Method 
- helps compute an apt value for k (num of clusters)
- k is initialized to a random val
- number of clusters can be high/low depending on the feature set and number of samples
- SSE = sum of square error used
  
  
- good k number based on this graph would be 4

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("income.csv")
df.head()
```

```python
plt.scatter(df["Age"], df["Income($)"])
```

```python
# based on the scatter plot made by the independent variables ---> k = 3 is logical

kmeans = KMeans(n_clusters=3)
kmeans
```

```python
# since it is an unsupervised learning algo ---> fit_predict instead of fit

y_predicted = kmeans.fit_predict(df[["Age", "Income($)"]])
y_predicted
```

```python
df["cluster"] = y_predicted
df.head()
```

```python
# 3 different data frames for the 3 different clusters.

df0 = df[df["cluster"]==0]
df1 = df[df["cluster"]==1]
df2 = df[df["cluster"]==2]

# plotting scatterplots for each cluster

plt.scatter(df0["Age"], df0["Income($)"], color='r')
plt.scatter(df1["Age"], df1["Income($)"], color='g')
plt.scatter(df2["Age"], df2["Income($)"], color='b')

plt.xlabel("Age")
plt.ylabel("Income($)")
```
