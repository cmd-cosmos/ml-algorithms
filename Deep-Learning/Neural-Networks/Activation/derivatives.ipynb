```python
#### calc. derivatives and plotting slopes 

import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
'''
1st derivative:

f'x = (f(x+h) - f(x)) / h
'''

def f(x):
    return (2*x**2 + 10)

print(f(10))
print(f(1))
print(f(0))
print(f(-90))
#### parabolic function 
```

```python
x_pts = np.arange(-10,10,1)
y_pts = [f(i) for i in x_pts]
print(x_pts)
print(y_pts)
```

```python
#### Plot 

plt.plot(x_pts, y_pts)
```

```python
#### function:    2*x**2 + 10 
#### derivative:  4x

def derivate(x,h):
    return (f(x+h) - f(x)) / h

x = 10
h = 0.001

derivate(x,h) # essentially 4x ---> expected 40
```
