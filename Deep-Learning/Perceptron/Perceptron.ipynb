```python
### Single unit of logic in an artificial NN
### algo that outputs a binary conclusion
### single node to mimic a neuron
### inputs ---> weights(determine contribution of input to decision) ---> output
### input---> weights*inputs---> weighted sum for all inp and weights
### weighed sum compared to threshold/bias ---> bias usually added during summation
### activation function used to determine if threashold reached or not
### if sum > threshold(hyperparameter) ---> return True
### hyperparameter ---> threshold/bias and initial weights
```

```python
import numpy as np

x_inputs: list = [0.2, 0.7, 0.5]
wts_init: list = [0.4, 0.9, 0.2]
threshold: float = 0.5

weighed_sum: float = 0.0

def step_activation_fn(weighed_sum:float) -> int:
    if weighed_sum > threshold:
        return 1
    else:
        return 0

def perceptron(weighed_sum: float) -> int:
    for inp,wt in zip(x_inputs, wts_init):
        weighed_sum += (inp*wt)
        print("current weighted sum: {:.3f}".format(weighed_sum))
    print("final weighted Sum: {:.3f}".format(weighed_sum))
    return step_activation_fn(weighed_sum)

decision = perceptron(weighed_sum)
print("perceptron decision = {:.3f}".format(decision))
```

```python

class Perceptron:
    '''
    Single Perceptron
    Activation Used: Step Activation Function
    '''

    def __init__(self, x_in, weights, threshold):
        self.x_in = x_in
        self.weights = weights
        self.threshold = threshold
        self.wt_sum = 0

    def activate(self):
            if self.wt_sum > threshold:
                return 1
            else:
                return 0

    def decider(self):
        for inp,wt in zip(self.x_in, self.weights):
            self.wt_sum += (inp*wt)
            print("current weighted sum: {:.3f}".format(self.wt_sum))
        print("final weighted Sum: {:.3f}".format(self.wt_sum))
        return self.activate()

x_inputs: list = [0.2, 0.7, 0.5]
wts_init: list = [0.4, 0.9, 0.2]
threshold: float = 0.5

perceptron = Perceptron(x_inputs, wts_init, threshold)
dec = perceptron.decider()
print("Decision: {}".format(dec))
```
