```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('heart.csv')
df.head()
```

```python
df.describe()
```

```python
cols_variance = ['Age', 'RestingBP', 'Cholesterol', 'FastingBS', 'MaxHR', 'Oldpeak', 'HeartDisease']

for col in cols_variance:
    df = df[(df[col] > df[col].mean() - 3 * df[col].std()) & (df[col] < df[col].mean() + 3 * df[col].std())]
df.head()
```

```python
df.describe()
```

```python
print(df['Sex'].unique())
print(df['ChestPainType'].unique())
print(df['RestingECG'].unique())
print(df['ExerciseAngina'].unique())
print(df['ST_Slope'].unique())
```

```python
df2 = df.copy()

df2['RestingECG'].replace(
    {
        'Normal' : 1,
        'ST' : 2,
        'LVH' : 3
    },
    inplace=True
)

df2['ExerciseAngina'].replace(
    {
        'N' : 0,
        'Y' : 1
    },
    inplace=True
)

df2['ST_Slope'].replace(
    {
        'Down' : 1,
        'Flat' : 2,
        'Up' : 3
    },
    inplace=True
)

df2.head()
```

```python
df_final = pd.get_dummies(df2, drop_first=True)
df_final.head()
```

```python
X = df_final.drop(['HeartDisease'], axis=1)
y = df_final['HeartDisease']

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

print(X_scaled)
print(np.shape(X_scaled))
print(X)
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=10)
```

```python
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
rfc.score(X_test, y_test)
```

```python
#### PCA

pca = PCA(0.95)
X_pca = pca.fit_transform(X)
X_pca
```

```python
pca_X_train, pca_X_test, pca_y_train, pca_y_test = train_test_split(X_pca, y, test_size=0.3, random_state=10)
print(np.shape(pca_X_train))
print(np.shape(pca_y_train))
```

```python
rfc_pca = RandomForestClassifier()
rfc_pca.fit(pca_X_train, pca_y_train)
rfc_pca.score(pca_X_test, pca_y_test)
```
