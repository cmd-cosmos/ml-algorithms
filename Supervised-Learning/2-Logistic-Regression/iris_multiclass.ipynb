```python
### multiclass predictions for the iris dataset

%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn import datasets
```

```python
iris_dataset = datasets.load_iris()
# dir method lists all attributes and methods in the scope passed as argument to the function
dir(iris_dataset)
```

```python
print(iris_dataset.DESCR)
```

```python

for i in range(len(iris_dataset.data)):
    print(iris_dataset.data[i])
    print(iris_dataset.target[i])
    print(iris_dataset.target_names)
    print('\n')
```

```python
# sepal length is the first attribute and width is the 2nd feature
fig,ax = plt.subplots()
ax.scatter(iris_dataset.data[:, 0], iris_dataset.data[:, 1], c=iris_dataset.target)
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
ax.legend(scatter.legend_elements()[0], iris_dataset.target_names)
plt.show()
```

```python
# train test split
X = iris_dataset.data
y = iris_dataset.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# model training

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test) ### 100% percent accuracy
```

```python
y_predicted = model.predict(X_test)

conf_matrix = confusion_matrix(y_test, y_predicted)
conf_matrix
```

```python
class_names = iris_dataset.target_names

sns.heatmap(conf_matrix, annot=True, xticklabels=class_names, yticklabels=class_names)
plt.xlabel("Predicted")
plt.ylabel("True")
```
