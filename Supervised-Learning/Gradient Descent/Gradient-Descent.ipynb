```python
'''
Author: Premanshu Bhagat
Date: 25/6/2025

Manual implementation of gradient descent.
'''

'''
- given labeled training data
- assume a linear regression problem
- find best fit line
- linear regression - viable approach
- y = mx + b, take random (m,b) and plot a line
- calcluate the error between predicted value and label/true value
- take square of the error and sum them up ----> MSE ----> Mean Square Error
- square to prevent skewing of result as errors can be -ve

- MSE is also called the cost function as it is a summation of individual losses
- Gradient descent is the best optimization function for suh a condition
- less iterations could result in convergence/global optimum and that m,b will be used for prediction
- risk of overshooting and undershooting is affected by the learning rate
- to ensure correct steps ---> take slope at each point ---> steepest descent plus learning rate to ensure restricted motion
- 

'''
```

```python
import numpy as np
import matplotlib.pyplot as plt


def gradient_descent(x: np.array ,y: np.array):
    m_current: int = 0 
    b_current: int = 0
    max_iterations: int = 1000
    alpha: float = 0.01  # learning rate ---> hyperparameter
    n: int = len(x)
    plt.title('Visualizing Gradient Descent')
    plt.xlabel('x')
    plt.ylabel('y_hat (predicted y)')
    plt.scatter(x,y,color='r')
    for i in range(max_iterations):
        # predictions
        y_hat = m_current*x + b_current # y = mx + b
        if (i!=max_iterations):
            plt.plot(x,y_hat,color='y',linestyle='solid')
        #cost ---> MSE
        cost = (1/n)*sum((y-y_hat)**2)
        # take partial derivatives to get direction of descent
        pd_m = -(2/n)*sum(x*(y-y_hat))  # partial derivative wrt m
        pd_b = -(2/n)*sum(y-y_hat)      # partial derivative wrt b
        # update m and b to get closer to optimum/minimum
        m_current -= (alpha*pd_m)
        b_current -= (alpha*pd_b)
        print("m: %.4f, b: %.4f, cost: %.4f, iteration: %d" %(m_current,b_current,cost,i))
    # final line after convergence ---> after optimization of (m,b)
    plt.plot(x,y_hat,color='g', linestyle='dashed')
    
x = np.array([1,2,3,4,5])
y = np.array([6,7,8,9,10])

gradient_descent(x,y)
# cost reduced down to 0.0040
```
