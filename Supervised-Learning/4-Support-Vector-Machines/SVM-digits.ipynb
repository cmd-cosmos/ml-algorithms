```python
### using the SVC to classify images from the digits dataset.

%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
digits = load_digits()
dir(digits)
```

```python
digits.images[0]
plt.gray()
plt.imshow(digits.images[100])
```

```python
digits.data
```

```python
digits.target_names
```

```python
df = pd.DataFrame(digits.data)
df["target"] = digits.target
df.head()
```

```python
X = df.drop(["target"], axis=1)
y = df.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)
```

```python
# training an SVC with the rbf kernel

rbf_svc = SVC()
rbf_svc.fit(X,y)
```

```python
# accuracy score of 99.81%

rbf_svc.score(X_test, y_test)
```

```python
# training an SVC with a linear kernel

linear_svc = SVC(kernel='linear')
linear_svc.fit(X_train, y_train)
```

```python
# accuracy score for the SVC with a linear kernel ---> 98.14%

linear_svc.score(X_test, y_test)
```

```python
# prediciton by the SVC with the linear kernel

plt.gray()
plt.imshow(digits.images[123])

linear_prediction_123 = linear_svc.predict([digits.data[123]])
print(linear_prediction_123)
```

```python
# prediciton by the SVC with the rbf kernel

plt.gray()
plt.imshow(digits.images[21])

rbf_prediction_21 = rbf_svc.predict([digits.data[21]])
print(rbf_prediction_21)
```
