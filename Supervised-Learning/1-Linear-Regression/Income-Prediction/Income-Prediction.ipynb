```python
'''
Author: Premanshu Bhagat
Date: 23/06/2025
Aim: employ linear regression model to predict per capita income in Canada given the year
'''

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model

```

```python
df = pd.read_csv('canada_per_capita_income.csv')
print(df.head(15)) # first 5 values
income = df['per capita income (US$)']
```

```python
plt.figure()
plt.title('CSV data Visualized')
plt.xlabel('Year')
plt.ylabel('Income USD')
plt.scatter(df['year'], df['per capita income (US$)'])
plt.show()
```

```python
reg = linear_model.LinearRegression() # linear regression object
reg.fit(df[['year']],income)
```

```python
pred_val = reg.predict([[2020]])
print(f"Predicted Value = {pred_val}")
```

```python
plt.scatter(df['year'], df['per capita income (US$)'])
plt.plot(df['year'], reg.predict(df[['year']]), color='red')

plt.title('Actual vs Linear Regression')
plt.xlabel('Year')
plt.ylabel('Income (US$)')
plt.show()
```
